##### Display the color used for each corresponding segmentation case

```python
segmentation_mask_colors = pd.read_csv(colors)
segmentation_mask_colors.head()

segmentation_color_index = 0
fig, axis = plt.subplots(6,4)
fig.set_figwidth(20)
fig.set_figheight(24)
fig.suptitle('Segmentation colors', fontsize=40)

color_num = 0
for i in range(0,6):
        for j in range(0,4):
                axis[i,j].set_facecolor((segmentation_mask_colors.loc[segmentation_color_index,'r']/255,segmentation_mask_colors.loc[segmentation_color_index,'g']/255,segmentation_mask_colors.loc[segmentation_color_index,'b']/255))
                axis[i,j].set_title(segmentation_mask_colors.loc[segmentation_color_index, 'name'])
                axis[i,j].tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
                segmentation_color_index +=1
plt.show()

dataset.plot_examples()
```
